```python
import numpy as np

def scratch_train_test_split(x,y, train_size=0.8, seed=0):
    """
        x : The following forms of ndarray, shape (n_samples, n_features)
        Training data
        
    y : The following forms of ndarray, shape (n_samples,)
        correct answer value
    train_size : float (0<train_size<1)
        specify what percentage to use as a train
    Returns
    x_train : The following forms of ndarray, shape (n_samples, n_features)
        Training data
    x_test : The following forms of ndarray, shape(n_saples, n_features)
        validation data
    y_train : The following forms of ndarray, shape (n_samples,)
        correct data value of training data
    y_test : The following forms of ndarray, shape (n_samples,)
        correct value of verification data
    """

        
        
    np.random.seed(seed)
    np.random.shuffle(x)
    
    np.random.seed(seed)
    np.random.shuffle(y)
    
    elements_x = len(x)
    elements_y = len(y)
    
    x_split = int(elements_x* train_size)
    y_split = int(elements_y* train_size)
    x_train, x_test = x[:x_split], x[x_split:]
    y_train, y_test = y[:y_split], y[y_split:]
    
    return x_train, x_test, y_train, y_test
```

```python
x_train, x_test, y_train, y_test = scratch_train_test_split(x,y, train_size=0.8)
print(x_train)
print(y_train)
```
